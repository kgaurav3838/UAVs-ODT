## Types of numbers

Python has various "types" of numbers (numeric literals). We'll mainly focus on integers and floating point numbers.

Integers are just whole numbers, positive or negative. For example: 2 and -2 are examples of integers.

```python
# Array 1-D
import numpy as np
arr0 = np.array([1,2,3,4,5,6], dtype = np.int16) # By default int32 or uint32
arr01=np.array([(1),(2),(3),(4),(5),(6)])
[x,y,z] = np.split(arr0,3) # other splits are dsplit, hsplit, vsplit
a1 = np.roll(arr0,2) # circular shift by 2
print('arr0=\n',arr0,'\narr0.shape=',arr0.shape,'\narr01=\n',arr01,'\narr01.shape=',arr01.shape)
print(arr0.dtype, arr01.dtype) # array dtype
print('Split arr0 (x,y,z)=',x,y,z) 
print('roll by 2=',a1)
# short trick to create a list containing integers from 0 to 10
x1 = [*range(10)] ; y1 = [*range(2,10,2)]
print('x1=',x1,'\ny1=',y1)
```

```python
# Array 2-D
# Below we have array representation in two different ways, say array-1 and array-2
import numpy as np
arr1=np.array([(1,2),(3,4)])  # Array-1
print('arr1=\n',arr1,'\narr1.shape=',arr1.shape,'and arr1 indices=',arr1[0,0],arr1[0,1],arr1[1,0],arr1[1,1])

arr2=np.array([[1,2],[3,4]]) # Array-2
print('\narr2=\n',arr2,'\narr2.shape=',arr2.shape,'and arr2 indices=',arr2[0,0],arr2[0,1],arr2[1,0],arr2[1,1])

# Flip: axis=0 (Horizontally); axis=1 or -1 (vertically)
x1 = np.flip(arr1) ; x2 = np.flip(arr1, axis=1)
print("Flip arr1 mirror=\n",x1,'\n Flip arr1 vertically=\n',x2)
a2 = np.rot90(arr1) # Rotate arr1 by 90 degree anticlockwise
print('Rotate arr1 by 90 degree=',a2)
```

```python
# Array 3-D
# One more extra square brackets will be seen in 3-D matrices representation compare to 2-D.
arr3=np.array([(1,2),(3,4)]).reshape(-1,1,2) # Create 3rd dimension; (a,b) dimesion becomes (a,1,b) if reshape(-1,1,2)
# arr=np.array([(1,2),(3,4)]).reshape(2,1,2) # Same as above
print('arr3=\n',arr3,'\narr3.shape=\n',arr3.shape,'and arr3 indices=',arr3[0,0,0],arr3[0,0,1],arr3[1,0,0],arr3[1,0,1])
arr4 = np.array([(1,2),(3,4),(5,6)]).reshape(-1,1,2) # Create shape=(3,1,2)
print('\narr4=\n',arr4,'\narr4.shape=',arr4.shape,'and arr4 indices=',arr4[0,0,0],arr4[0,0,1],arr4[1,0,0],arr4[1,0,1],arr4[2,0,0],arr4[2,0,1])
```

```python
# numpy vs list speed of operation
import numpy as np
import sys
import time
size=100000
L1=range(size); L2=range(size)
N1=np.arange(size); N2=np.arange(size)

#Adding two list L1 and L2
start=time.time()
result= [(x+y) for x,y in zip(L1,L2)]  # Need for loop in order to add L1 and L2
print('Time taken by list operations=',(time.time()-start)*1000) # multiply 1000 to convert into milli sec

#Adding two nummpy  N1 and N2
start=time.time()
result=N1 + N2
print('Time taken by numpy operations=',(time.time()-start)*1000) #printing time taken by numpy operations
print('result',result)
```

# Bitwise operations on Numpy Array

```python
import numpy as np
x = np.array([0, 1, 0, 1])
y = np.array([0, 0, 1, 1],np.uint8)

print(np.bitwise_and(x, y))
print(np.bitwise_or(x, y))
print(np.bitwise_xor(x, y))
print(np.bitwise_not(x))
print(np.bitwise_not(y))
```

# FFT and IFFT

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
# 1-D FFT
t = np.arange(128)
# print(t)
sp = np.fft.fft(np.sin(t))  # FFT
freq = np.fft.fftfreq(t.shape[-1])
plt.subplot( 2, 1, 1)
plt.title('Real Part[FFT]')
plt.plot(freq, sp.real)
plt.subplots_adjust(hspace=0.5)
plt.subplot( 2, 1, 2)
plt.title('Imaginary Part[FFT]')
plt.plot(freq, sp.imag)
plt.show()
# Inverse 1-D DFT
t = np.arange(500)
n = np.zeros((500, ),dtype = complex)
n[40:80] = np.exp(1j*np.random.uniform(0, 2*np.pi, (40,)))
s = np.fft.ifft(n)
#print(t, n)
plt.subplot( 2, 1, 1)
plt.title('Real Part[IFFT]')
plt.plot(t, s.real)
plt.subplots_adjust(hspace=0.5)
plt.subplot( 2, 1, 2)
plt.title('Imaginary Part[IFFT]')
plt.plot(t, s.imag)
plt.show()
```
