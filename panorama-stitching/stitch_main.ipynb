```python
# USAGE
# python stitch

import cv2
import imutils
import time
import sys
import matplotlib.pyplot as plt
start=time.time()
plt.rcParams['figure.figsize'] = (30,30)
# import the necessary packages
from imagesearch.panorama import Stitcher
def show(img,i,title:str):
    plt.subplot(4,1,i)
    plt.imshow(img)
    plt.title(title)
# load the two images and resize them to have a width of 400 pixels
# (for faster processing)
imageA = cv2.imread("images/left1.png")
imageB = cv2.imread("images/right1.png")
imageA = imutils.resize(imageA, width=800)
imageB = imutils.resize(imageB, width=800)

# stitch the images together to create a panorama
stitcher = Stitcher()
(result, vis) = stitcher.stitch([imageA[150:270,:], imageB[150:270,:]], showMatches=True)
   
show(imageA,1,"imgA")
show(imageB,2,"imgB")
show(vis,3,"Keypoints")
show(result,4,"result")

print("Operational time=",(time.time()-start)*1000)
```
